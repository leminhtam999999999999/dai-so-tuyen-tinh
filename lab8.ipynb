```python
#Bài 2
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Định nghĩa các biến ký hiệu
x1, x2, y1, y2, c = sp.symbols('x1 x2 y1 y2 c')

# Ánh xạ 1: f1(x, y) = (2x + y, x - y) [Dự kiến là tuyến tính]
def f1(x, y):
    return (2*x + y, x - y)

# Ánh xạ 2: f2(x, y) = (x + y, x^2) [Dự kiến không tuyến tính]
def f2(x, y):
    return (x + y, x**2)

# Hàm kiểm tra tính tuyến tính
def kiem_tra_tuyen_tinh(f, ten):
    print(f"\nKiểm tra tính tuyến tính cho {ten}:")

    # Kiểm tra tính cộng: f(u + v) = f(u) + f(v)
    u = (x1, x2)
    v = (y1, y2)
    f_u_cong_v = f(x1 + y1, x2 + y2)
    f_u = f(x1, x2)
    f_v = f(y1, y2)
    f_u_cong_f_v = (f_u[0] + f_v[0], f_u[1] + f_v[1])
    tinh_cong = f_u_cong_v == f_u_cong_f_v
    print("Tính cộng đúng?", tinh_cong)
    print("f(u + v) =", f_u_cong_v)
    print("f(u) + f(v) =", f_u_cong_f_v)

    # Kiểm tra tính đồng nhất: f(cu) = cf(u)
    cu = (c*x1, c*x2)
    f_cu = f(c*x1, c*x2)
    cf_u = (c*f_u[0], c*f_u[1])
    tinh_dong_nhat = f_cu == cf_u
    print("Tính đồng nhất đúng?", tinh_dong_nhat)
    print("f(cu) =", f_cu)
    print("cf(u) =", cf_u)

    # Kết luận
    if tinh_cong and tinh_dong_nhat:
        print(f"Kết luận: {ten} là ánh xạ tuyến tính.")
    else:
        print(f"Kết luận: {ten} không phải là ánh xạ tuyến tính.")

# Chạy kiểm tra cho cả hai ánh xạ
kiem_tra_tuyen_tinh(f1, "f1(x, y) = (2x + y, x - y)")
kiem_tra_tuyen_tinh(f2, "f2(x, y) = (x + y, x^2)")

# Vẽ đồ thị minh họa
plt.figure(figsize=(12, 5))

# Các vector để biến đổi
vectors = [(1, 0), (0, 1), (1, 1)]

# Đồ thị cho f1
plt.subplot(1, 2, 1)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
for vec in vectors:
    x, y = vec
    fx, fy = f1(x, y)
    plt.arrow(0, 0, x, y, color='blue', width=0.02, head_width=0.1, alpha=0.5, label='Vector gốc' if vec == (1, 0) else "")
    plt.arrow(0, 0, fx, fy, color='red', width=0.01, head_width=0.08, label='Vector sau f1' if vec == (1, 0) else "")
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ánh xạ f1(x, y) = (2x + y, x - y)')
plt.legend()
plt.axis('equal')

# Đồ thị cho f2
plt.subplot(1, 2, 2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
for vec in vectors:
    x, y = vec
    fx, fy = f2(x, y)
    plt.arrow(0, 0, x, y, color='blue', width=0.02, head_width=0.1, alpha=0.5, label='Vector gốc' if vec == (1, 0) else "")
    plt.arrow(0, 0, fx, fy, color='green', width=0.01, head_width=0.08, label='Vector sau f2' if vec == (1, 0) else "")
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ánh xạ f2(x, y) = (x + y, x^2)')
plt.legend()
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
x, y, c = sp.symbols('x y c')
phuong_trinh = x**2 + y**2 - 1
def ve_do_thi_elip(a, b, ten_hinh):
    theta = np.linspace(0, 2*np.pi, 100)
    x_vals = a * np.cos(theta)
    y_vals = b * np.sin(theta)
    plt.figure(figsize=(6, 6))
    plt.plot(x_vals, y_vals, label=ten_hinh, color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Đồ thị {ten_hinh}')
    plt.legend()
    plt.axis('equal')
    plt.show()
ve_do_thi_elip(1, 1, 'Elip (x^2 + y^2 - 1 = 0)')

# Hàm ánh xạ xoay 45 độ
def anh_xa_xoay_sym(x, y):
    cos_theta = sp.sqrt(2) / 2  # cos(45°) = √2/2
    sin_theta = sp.sqrt(2) / 2  # sin(45°) = √2/2
    mat_xoay = sp.Matrix([[cos_theta, -sin_theta],
                          [sin_theta, cos_theta]])
    vector = sp.Matrix([x, y])
    return mat_xoay * vector


def kiem_tra_tuyen_tinh(f, ten):
    print(f"\nKiểm tra tính tuyến tính cho {ten}:")
    u = sp.Matrix([x, y])
    v = sp.Matrix([x, y])
    f_u_cong_v = f(x + x, y + y)
    f_u = f(x, y)
    f_v = f(x, y)
    f_u_cong_f_v = f_u + f_v
    tinh_cong = f_u_cong_v == f_u_cong_f_v
    print("Tính cộng đúng?", tinh_cong)
    print("f(u + v) =", f_u_cong_v)
    print("f(u) + f(v) =", f_u_cong_f_v)

    # Kiểm tra tính đồng nhất: f(cu) = cf(u)
    cu = c * u
    f_cu = f(c * x, c * y)
    cf_u = c * f_u
    tinh_dong_nhat = f_cu == cf_u
    print("Tính đồng nhất đúng?", tinh_dong_nhat)
    print("f(cu) =", f_cu)
    print("cf(u) =", cf_u)

    if tinh_cong and tinh_dong_nhat:
        print(f"Kết luận: {ten} là ánh xạ tuyến tính.")
    else:
        print(f"Kết luận: {ten} không phải là ánh xạ tuyến tính.")
kiem_tra_tuyen_tinh(anh_xa_xoay_sym, "Ánh xạ xoay 45 độ")
def anh_xa_xoay_num(x, y, theta=np.pi/4):
    mat_xoay = np.array([[np.cos(theta), -np.sin(theta)],
                         [np.sin(theta), np.cos(theta)]])
    vector = np.array([x, y])
    return mat_xoay @ vector

vectors = [(1, 0), (0, 1), (1, 1)]
plt.figure(figsize=(6, 6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
for vec in vectors:
    x, y = vec
    xoay_vec = anh_xa_xoay_num(x, y)
    plt.arrow(0, 0, x, y, color='blue', width=0.02, head_width=0.1, alpha=0.5, label='Vector gốc' if vec == (1, 0) else "")
    plt.arrow(0, 0, xoay_vec[0], xoay_vec[1], color='red', width=0.01, head_width=0.08, label='Vector sau xoay' if vec == (1, 0) else "")
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ánh xạ xoay 45 độ')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
from numpy import linalg as LA

A = np.matrix([[1,0],[0,0]])
B = np.matrix([[0,0],[0,1]])

try:
    LA.inv(A)
except LA.LinAlgError as e:
    print("LA.inv(A)", str(e))

try:
    LA.inv(B)
except LA.LinAlgError as e:
    print("LA.inv(B)", str(e))

try:
    LA.inv(A+B)
except LA.LinAlgError as e:
    print("LA.inv(A+B)", str(e))
else:
    print("LA.inv(A+B)", LA.inv(A+B))

print("LA.det(A)", LA.det(A))
print("LA.det(B)", LA.det(B))
print("LA.det(A+B)", LA.det(A+B))
```

```python
import numpy as np
from numpy import linalg as LA

u = np.array([1, 2, 3])
v = np.array([0, 1, 2])
w = np.array([-1, 0, 1])

# Phần a: Kiểm tra (1,1,1) là tổ hợp tuyến tính
A = np.column_stack((u, v, w))
b_a = np.array([1, 1, 1])
x_a = LA.solve(A, b_a)
print("Phần a: Hệ số cho (1,1,1) =", x_a)

# Phần b: Kiểm tra (1,-2,2) không là tổ hợp tuyến tính
b_b = np.array([1, -2, 2])
try:
    x_b = LA.solve(A, b_b)
except LA.LinAlgError:
    print("Phần b: (1,-2,2) không là tổ hợp tuyến tính (hệ không có nghiệm duy nhất)")
```
